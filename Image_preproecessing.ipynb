```python
coins = data.coins()

print('Type',type(coins))
print('dtype', coins.dtype)
print('Shape', coins.shape)

plt.imshow(coins)
```

```python
cat = data.chelsea()
print('Shape', cat.shape)
print('Values min and max',cat.min(), cat.max())

plt.imshow(cat)
```
